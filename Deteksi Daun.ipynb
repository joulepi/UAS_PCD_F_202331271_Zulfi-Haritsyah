## Import Library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Baca Citra Asli

```python
img = cv2.imread('buahdaun1.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## Mask Buah

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

lower_buah = np.array([15, 80, 80])
upper_buah = np.array([35, 255, 255])
mask_buah = cv2.inRange(hsv, lower_buah, upper_buah)
```

## Mask Daun

```python
lower_daun = np.array([35, 40, 40])
upper_daun = np.array([85, 255,255])
mask_daun = cv2.inRange(hsv, lower_daun, upper_daun)
```

## Segmentasi Buah

```python
seg_buah = cv2.bitwise_and(img_rgb, img_rgb, mask=mask_buah)
```

## Segmentasi Daun

```python
seg_daun = cv2.bitwise_and(img_rgb, img_rgb, mask=mask_daun)
```

## Tampilkan Semuanya

```python
fig, axs = plt.subplots(1, 4, figsize=(15,5))

axs[0].imshow(img_rgb)
axs[0].set_title('Gambar Asli')
axs[0].axis('off')

axs[1].imshow(mask_buah)
axs[1].set_title('Mask Buah')
axs[1].axis('off')

axs[2].imshow(seg_buah)
axs[2].set_title('Segmentasi Buah')
axs[2].axis('off')

axs[3].imshow(seg_daun)
axs[3].set_title('Segmentasi Daun')
axs[3].axis('off')

plt.tight_layout()
plt.show()
```
