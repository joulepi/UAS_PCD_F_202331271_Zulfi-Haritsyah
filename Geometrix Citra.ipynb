## Import Library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Citra Asli

```python
img = cv2.imread("diri.jpg")  
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## Rotated

```python
(h, w) = img_rgb.shape[:2]
center = (w // 2, h // 2)
M_rot = cv2.getRotationMatrix2D(center, 30, 1.0)  # Rotasi 30 derajat
rotated = cv2.warpAffine(img_rgb, M_rot, (w, h))
```

## Resized

```python
resized = cv2.resize(img_rgb, (1000, 1200)) 
```

## Cropped

```python
cropped = img_rgb[900:2200, 500:1500]
```

## Flipped

```python
flipped = cv2.flip(img_rgb, 1) # Horizontal flip (Mirror)
```

## Translated

```python
M_trans = np.float32([[1, 0, 300], [0, 1, 200]]) # Geser kanan 300px, bawah 200px
translated = cv2.warpAffine(img_rgb, M_trans, (w, h))
```

## Tampilkan Hasil

```python
titles = ["Citra Asli", "After Rotated", "After Resized", "After Cropped", "After Flipped", "After Translated"]
images = [img_rgb, rotated, resized, cropped, flipped, translated]

plt.figure(figsize=(15,10))
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("on")
plt.tight_layout()
plt.show()
```
